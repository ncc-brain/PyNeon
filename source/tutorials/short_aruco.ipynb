```python
from pyneon import Dataset, get_sample_data
dataset_dir = get_sample_data("markers", format="cloud")
dataset = Dataset(dataset_dir)
rec = dataset.recordings[1]
video = rec.scene_video
```

## Marker detection every frame

```python
detected_markers = video.detect_markers("5x5_100")
```

```python
video.plot_detected_markers(detected_markers, frame_index=498)
```

```python
video.overlay_detected_markers(
    detected_markers,
    show_video=True,
)
```

## Marker detection with steps

```python
detected_markers_steps = video.detect_markers("5x5_100", step=3)
```

```python
video.plot_detected_markers(detected_markers_steps, frame_index=498)
```

```python
video.overlay_detected_markers(
    detected_markers_steps,
    show_video=True,
)
```
