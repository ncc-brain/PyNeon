# Export Neon data recording to BIDS formats


```python
from pyneon import get_sample_data, NeonRecording

sample_dir = get_sample_data("OfficeWalk")
recording_dir = sample_dir / "Timeseries Data" / "walk1-e116e606"
recording = NeonRecording(recording_dir)
motion_dir = sample_dir / "BIDS" / "sub-1" / "motion"
motion_dir.mkdir(exist_ok=True, parents=True)
```

```python
recording.to_motion_bids(
    motion_dir, prefix="sub-1_task-officeWalk_tracksys-NeonIMU_run-1"
)
```
